# Soru 3: Musteri Yönetim Sistemi

      # kullanındepolamak , yeni musterileri seçmek , musterileri ​
 Bu proje, musterilerinizi yönlendirmek ve temel islemleri gerceklestirmek için kullanabileceginiz
    bir sistemi musteri yonetim olusturmanizi icerir. Bu sistem, musterileri depolamak, yeni musterileri seçmek, 
    musterileri toplamak, musterileri silmek ve musteri listesini görmek gibi temel islevlere sahip olacaktır.


# Projenin temel adımları sunlardır:

 1- 1-Müşteri bilgilerini depolamak için bir sözlük yapısı kullanın.. Her müşteri için benzersiz bir müşteri kimliği (ID) atayın ve
 müşteri bilgilerini bu kimlikle ilişkilendirin.Her müşteri için 
 ad, soyad, e-posta, telefon numarası gibi bilgileri içeren bir sözlük kullanabilirsiniz.
 2-Kullanıcının aşağıdaki eylemleri seçebileceği bir menü sağlayın:


```python

import random
musteriler = {
    1001: ["Ahmet",  "Yilmaz", "ahmet.yilmaz@example.com",  "05001234567"],
    1002: ["Ayse",   "Kara",   "ayse.kara@example.com",     "05559876543"],
    1003: ["Mehmet", "Demir",  "mehmet.demir@example.com",  "05331239876"],
    1004: ["Zeynep", "Aydin",  "zeynep.aydin@example.com",  "05449811223"],
    1005: ["Can",    "Kurt",   "can.kurt@example.com",      "05006789432"],
    1006: ["Elif",   "Yildiz", "elif.yildiz@example.com",   "05336781234"],
    1007: ["Burak",  "Aslan",  "burak.aslan@example.com",   "05227893456"],
    1008: ["Fatma",  "Celik",  "fatma.celik@example.com",   "05114567891"],
    1009: ["Emre",   "Sahin",  "emre.sahin@example.com",    "05448912345"],
    1010: ["Melis",  "Gunes",  "melis.gunes@example.com",   "05331238765"]
}

def yeni_musteriler_ekle():
    while True:
        anahtar = random.randint(1011,9999)
        if anahtar not in musteriler.keys():
            ad = input("yeni musterinin adini giriniz: ")
            soyad = input("yeni musterinin soyadini giriniz: ")
            email = input("yeni musterinin email adresini giriniz: ")
            telefon = input("yeni musterinin telefon numarasini giriniz: ")
            musteriler[anahtar] = [ad,soyad,email,telefon]
            break
def musteri_bilgilerinin_güncellenmesi():
    guncellenecek_musteri = int(input("guncellenmesini istediginiz musterinin id'sini giriniz"))
    musteri_secenek = input("musterinin neyini guncellemek istiyorsunuz(ad,soyad,email,tel)").lower()
    if musteri_secenek == "ad":
        musteriler[guncellenecek_musteri][0] = input("yeni musterinin adini giriniz")
    elif musteri_secenek == "soyad":
        musteriler[guncellenecek_musteri][1] = input("yeni musterinin soyadini giriniz")
    elif musteri_secenek == "email":
        musteriler[guncellenecek_musteri][2] = input("yeni musterinin email giriniz")
    elif musteri_secenek == "tel":
        musteriler[guncellenecek_musteri][3] = input("yeni musterinin telefon giriniz")
    else:
        print("yanlis giris")
def musteriyi_sil():
    silinecek_musteri = int(input("silinmesini istedigini musterinin id'sini giriniz"))
    del musteriler[silinecek_musteri]
def tum_musterileri_listele():
    print(musteriler,sep="\n",end="\n")


while True:
    kullanici_girisi = int(input("""
1 - Yeni musteriler ekle
2 - Musteri bilgilerinin guncellenmesi                 
3 - Musteriyi sil  
4 - Tum musterileri listele
5 - Cikis yapmak
"""))

    if kullanici_girisi == 1:
        yeni_musteriler_ekle()
    elif kullanici_girisi == 2:
        musteri_bilgilerinin_güncellenmesi()
    elif kullanici_girisi == 3:
        musteriyi_sil()
    elif kullanici_girisi == 4:
        tum_musterileri_listele()
    elif kullanici_girisi == 5:
        break
    else:
        print("hatali giris yaptiniz.... ")

    
    
  
```
